```python
import torch

x = torch.arange(4.0)
x
```

```python
x.requires_grad_(True)
x.grad
```

```python
y = 2 * torch.dot(x,x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad == 4 * x
```

```python
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```

```python
x.grad.zero_()
y = x * x
print(y)
y.sum()
print(y.sum())
y.sum().backward()
# print(y.sum().backward())
print(x.grad)
```
